```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler, StandardScaler,RobustScaler
from sklearn.metrics import classification_report
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
data = pd.read_csv('Housing.csv')
df = data
```

```python
df
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df['HighPrice'] = (df['price']>5000000).astype(int)
```

```python
df
```

```python
#feature selectiom and target
x = df[['area','bedrooms','bathrooms','stories']]
y = df['HighPrice'] 
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y, test_size = 0.2, random_state = 35)
```

```python
scaler = MinMaxScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_train_scaled
```

```python
x_test_scaled = scaler.fit_transform(x_test)
x_test_scaled
```

```python
x_train
```

```python
x_test
```

```python
model = LogisticRegression()
```

```python
#training model
model.fit(x_train, y_train)
```

```python
y_predict = model.predict(x_test)
y_predict
#predicted value based on test data
```

```python
#actal value present in out data
y_test
```

```python
#accuray for unscaled data
accuracy_oflr = accuracy_score(y_predict,y_test)
print(f'Acuracy of the model: {accuracy_oflr}')
```

```python
print(classification_report(y_test,y_predict))
```

```python
x_test
```

```python
#training model after scalling
model.fit(x_train_scaled,y_train)
```

```python
#prediction
y_predict_scaled = model.predict(x_test_scaled)
#elivating
#accuracy for scaled data 
accuracy_of_lr = accuracy_score(y_predict_scaled,y_test)
print(f'Acuuracy of the model: {np.round(accuracy_of_lr,2)}')
```

```python
#confusion Matrix
print(classification_report(y_test,y_predict_scaled))
```
